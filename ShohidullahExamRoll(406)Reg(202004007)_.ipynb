## Conditional Statements
Conditional statements allow to execute certain blocks of code based on whether a condition is true or false. They are essential for decision-making in programming.

**Types of Conditional Statements**
1. if statement
2. if-else statement
3. if-elif-else statement
4. Nested if statements

### `if` Statement
The statement evaluates a condition and executes the indented block of code if the condition is true.

```
if condition:
    # code block
```

````python
age = 18

if age >= 18:
    print("You are an adult.")
````

### `if-else` Statement
The `else` statement is used to execute a block of code if the condition in the `if` statement is false.
```
if condition:
    # block of code if condition is true
else:
    # block of code if condition is false

```

````python
age = 16
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
````

### `elif` Statement
The `if-elif-else` statement allows checking multiple conditions. The first true condition's code block gets executed.

```
if condition1:
    # code block for condition1
elif condition2:
    # code block for condition2
else:
    # code block if none of the above conditions are true
```

````python
score = 75
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 50:
    print("Grade: D")
else:
    print("Grade: F")
````

### Nested `if` Statements
 `if`, `elif`, and `else` statements can be nested to create more complex conditions.

 ```
if condition1:
    if condition2:
        # block of code if condition1 and condition2 are true
    else:
        # block of code if condition1 is true but condition2 is false
else:
    # block of code if condition1 is false
 ```

````python
num = 10
if num > 0:
    print("The number is positive.")
    if num % 2 == 0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is negative or zero.")
````

````python
# =====================
# 1. Data Types (1–15)
# =====================

# 1. Input type check
x = input("Enter something: ")
print("Type:", type(x))

# 2. Variable type int
x = 10
print(type(x))

# 3. Variable type float
y = 5.5
print(type(y))

# 4. Variable type string
name = "Python"
print(type(name))

# 5. Boolean type
is_valid = True
print(type(is_valid))

# 6. Convert string to int
s = "100"
num = int(s)
print(num, type(num))

# 7. Convert int to float
x = 50
y = float(x)
print(y, type(y))

# 8. List type
lst = [1,2,3,4,5]
print(type(lst))

# 9. Tuple type
t = (1,2,3)
print(type(t))

# 10. Set type
s = {1,2,3,2}
print(type(s))

# 11. Dictionary type
d = {'a':1,'b':2}
print(type(d))

# 12. Input two numbers as string and sum
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Sum:", a+b)

# 13. Input number and check type
num = float(input("Enter number: "))
if num == int(num):
    print("Int")
else:
    print("Float")

# 14. Complex number
c = 3+5j
print(c, type(c))

# 15. Check if input is numeric
x = input("Enter: ")
print("Is numeric?", x.isnumeric())


# =====================
# 2. Numbers (16–30)
# =====================

# 16. Even or Odd
num = int(input("Enter number: "))
print("Even" if num % 2 == 0 else "Odd")

# 17. Positive, Negative, Zero
num = int(input("Enter number: "))
if num>0:
    print("Positive")
elif num<0:
    print("Negative")
else:
    print("Zero")

# 18. Square
n = int(input("Enter number: "))
print("Square:", n**2)

# 19. Cube
print("Cube:", n**3)

# 20. Sum, Difference, Product, Division
a = int(input("Enter a: "))
b = int(input("Enter b: "))
print("Sum:", a+b)
print("Difference:", a-b)
print("Product:", a*b)
print("Division:", a/b)

# 21. Factorial
n = int(input("Enter number: "))
fact = 1
for i in range(1,n+1):
    fact *= i
print("Factorial:", fact)

# 22. Divisible by 5
n = int(input("Enter number: "))
print("Divisible by 5" if n%5==0 else "Not divisible")

# 23. Divisible by 3 and 5
print("Divisible by 3 and 5" if n%3==0 and n%5==0 else "Not divisible by both")

# 24. Prime check
n = int(input("Enter number: "))
if n>1:
    for i in range(2,n):
        if n%i==0:
            print("Not prime")
            break
    else:
        print("Prime")
else:
    print("Not prime")

# 25. Largest of two
a = int(input("a: "))
b = int(input("b: "))
print("Largest:", max(a,b))

# 26. Largest of three
a = int(input("a: "))
b = int(input("b: "))
c = int(input("c: "))
print("Largest:", max(a,b,c))

# 27. Multiplication table
num = int(input("Number: "))
for i in range(1,11):
    print(num,"x",i,"=",num*i)

# 28. Sum of digits
num = int(input("Number: "))
sum_digits = 0
for d in str(num):
    sum_digits += int(d)
print("Sum of digits:", sum_digits)

# 29. Reverse number
num = int(input("Number: "))
rev = 0
temp = num
while temp>0:
    rev = rev*10 + temp%10
    temp//=10
print("Reversed:", rev)

# 30. Armstrong number
num = int(input("Number: "))
sum_arm = sum(int(d)**3 for d in str(num))
print("Armstrong" if sum_arm==num else "Not Armstrong")


# =====================
# 3. Strings (31–45)
# =====================

# 31. Print string
s = input("Enter string: ")
print(s)

# 32. Full name
first = input("First name: ")
last = input("Last name: ")
print("Full Name:", first+" "+last)

# 33. String length
s = input("String: ")
print("Length:", len(s))

# 34. Uppercase
print(s.upper())

# 35. Lowercase
print(s.lower())

# 36. Alternate characters
print(s[::2])

# 37. Reverse string
print(s[::-1])

# 38. Count vowels
count = sum(1 for ch in s.lower() if ch in 'aeiou')
print("Vowels:", count)

# 39. Palindrome check
print("Palindrome" if s==s[::-1] else "Not palindrome")

# 40. Substring check
sub = input("Substring: ")
print("Found" if sub in s else "Not found")

# 41. Replace word
old = input("Old word: ")
new = input("New word: ")
print(s.replace(old,new))

# 42. Remove spaces
print(s.replace(" ",""))

# 43. First and last character
print("First:", s[0], "Last:", s[-1])

# 44. Title case
print(s.title())

# 45. Startswith check
prefix = input("Prefix: ")
print("Yes" if s.startswith(prefix) else "No")


# =====================
# 4. Slicing (46–60)
# =====================

# 46. First 5 chars
print(s[:5])

# 47. Last 5 chars
print(s[-5:])

# 48. Index 2 to 6
print(s[2:7])

# 49. Remove first and last
print(s[1:-1])

# 50. Every 2nd char
print(s[::2])

# 51. Reverse using slicing
print(s[::-1])

# 52. List first 3
lst = [int(x) for x in input("List: ").split()]
print(lst[:3])

# 53. List last 3
print(lst[-3:])

# 54. List index 2 to 5
print(lst[2:6])

# 55. Alternate elements
print(lst[::2])

# 56. Tuple first 3
t = tuple(int(x) for x in input("Tuple: ").split())
print(t[:3])

# 57. Tuple last 2
print(t[-2:])

# 58. String odd indices
print(s[1::2])

# 59. String even indices
print(s[0::2])

# 60. Remove vowels
print("".join(ch for ch in s if ch.lower() not in 'aeiou'))


# =====================
# 5. Conditional Statements (61–75)
# =====================

# 61. Grade
marks = int(input("Marks: "))
if marks>=80:
    print("Grade A")
elif marks>=60:
    print("Grade B")
elif marks>=40:
    print("Grade C")
else:
    print("Grade F")

# 62. Divisible by 3
n = int(input("Number: "))
print("Yes" if n%3==0 else "No")

# 63. Divisible by 5
print("Yes" if n%5==0 else "No")

# 64. Divisible by 3 or 5
print("Yes" if n%3==0 or n%5==0 else "No")

# 65. Divisible by 3 and 5
print("Yes" if n%3==0 and n%5==0 else "No")

# 66. Prime check
if n>1:
    for i in range(2,n):
        if n%i==0:
            print("Not prime")
            break
    else:
        print("Prime")
else:
    print("Not prime")

# 67. Leap year
year = int(input("Year: "))
if (year%4==0 and year%100!=0) or (year%400==0):
    print("Leap Year")
else:
    print("Not Leap Year")

# 68. Largest of two
a = int(input("a: "))
b = int(input("b: "))
print(max(a,b))

# 69. Largest of three
c = int(input("c: "))
print(max(a,b,c))

# 70. Vowel or consonant
ch = input("Char: ").lower()
print("Vowel" if ch in 'aeiou' else "Consonant")

# 71. Even/Odd function
def check_even_odd(n):
    print("Even" if n%2==0 else "Odd")
check_even_odd(int(input("Number: ")))

# 72. Voting eligibility
age = int(input("Age: "))
print("Eligible" if age>=18 else "Not Eligible")

# 73. Pos/Neg/Zero
n = int(input("Number: "))
if n>0:
    print("Positive")
elif n<0:
    print("Negative")
else:
    print("Zero")

# 74. Ascending order of three numbers
a = int(input("a: "))
b = int(input("b: "))
c = int(input("c: "))
print("Ascending:", sorted([a,b,c]))

# 75. Divisible by 2,3,5
print("Divisible by 2,3,5" if n%2==0 or n%3==0 or n%5==0 else "Not divisible")


# =====================
# 6. Loops (76–90)
# =====================

# 76. Numbers 1 to 10
for i in range(1,11):
    print(i)

# 77. While 1 to 10
i=1
while i<=10:
    print(i)
    i+=1

# 78. Even 1 to 20
for i in range(1,21):
    if i%2==0: print(i)

# 79. Odd 1 to 20
for i in range(1,21):
    if i%2!=0: print(i)

# 80. Sum first n
n = int(input("n: "))
sum = 0
for i in range(1,n+1):
    sum+=i
print("Sum:", sum)

# 81. Sum first n even
n = int(input("n: "))
sum_even=0
for i in range(2,2*n+1,2):
    sum_even+=i
print("Sum even:", sum_even)

# 82. Sum first n odd
sum_odd=0
for i in range(1,2*n,2):
    sum_odd+=i
print("Sum odd:", sum_odd)

# 83. Multiplication table
num=int(input("Number: "))
for i in range(1,11):
    print(num,'x',i,'=',num*i)

# 84. Reverse 10 to 1
for i in range(10,0,-1):
    print(i)

# 85. Pattern
rows=4
for i in range(1,rows+1):
    print("*"*i)

# 86. Square pattern
n=int(input("Rows: "))
for i in range(n):
    print("*"*n)

# 87. Triangle pattern of numbers
n=int(input("Rows: "))
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end="")
    print()

# 88. Fibonacci series
n=int(input("Terms: "))
a,b=0,1
for i in range(n):
    print(a,end=" ")
    a,b=b,a+b
print()

# 89. Sum of digits
num=int(input("Number: "))
sum_d=0
for d in str(num):
    sum_d+=int(d)
print("Sum digits:", sum_d)

# 90. Count digits
num=int(input("Number: "))
count=0
while num>0:
    num//=10
    count+=1
print("Digits:", count)


# =====================
# 7. Functions (91–100)
# =====================

# 91. Simple greet
def greet():
    print("Hello Python")
greet()

# 92. Add two numbers
def add(a,b):
    print("Sum:",a+b)
add(10,20)

# 93. Square
def square(n):
    return n**2
print(square(5))

# 94. Factorial
def factorial(n):
    f=1
    for i in range(1,n+1):
        f*=i
    return f
print(factorial(5))

# 95. Even/Odd check
def even_odd(n):
    print("Even" if n%2==0 else "Odd")
even_odd(7)

# 96. Greet user default
def greet_user(name="Guest"):
    print("Hello",name)
greet_user("Rahim")
greet_user()

# 97. Sum and Difference
def sum_diff(a,b):
    return a+b,a-b
s,d=sum_diff(10,5)
print("Sum:",s,"Diff:",d)

# 98. Recursive factorial
def rec_fact(n):
    if n==0 or n==1: return 1
    return n*rec_fact(n-1)
print(rec_fact(5))

# 99. Reverse string
def reverse_str(s):
    return s[::-1]
print(reverse_str("Python"))

# 100. Count vowels
def count_vowels(s):
    return sum(1 for ch in s.lower() if ch in 'aeiou')
print(count_vowels("PythonProgramming"))
````
